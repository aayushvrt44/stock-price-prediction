```python
import pandas as pd
import numpy as np
```

```python
ticker=pd.read_csv('ticker.csv')
```

```python
sector=pd.read_csv('test.csv')
```

```python
ticker_arr=np.array(ticker)
sector_arr=np.array(sector)
```

```python
merge=pd.DataFrame()
```

```python
k=0
flag=0
for i in range ((sector_arr.shape)[0]):
    flag=0
    for j in range((ticker_arr.shape)[0]):
        try:
       
        
            ans=set(sector_arr[i][0].split()).intersection(ticker_arr[j][1].split())
            if len(ans)>0: 
                data={"Ticker":ticker_arr[j][0],"company_name":ticker_arr[j][1],"sector":sector_arr[i][1],"series":ticker_arr[j][2],"date of listing":ticker_arr[j][3],
                "paid up value":ticker_arr[j][4],"market lot":ticker_arr[j][5],"isin number":ticker_arr[j][6],"face value":ticker_arr[j][7]}
                temp_df=pd.DataFrame(data,index=[k])
                k=k+1
                merge=merge.append(temp_df)
                flag=1
                break
                
        except:
             break
    
```

```python
merge.to_csv('merge.shape2.csv')
```
