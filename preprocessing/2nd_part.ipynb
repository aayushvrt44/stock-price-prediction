```python
k=0
flag=0
for i in range ((sector_arr.shape)[0]):
    flag=0
    for j in range((ticker_arr.shape)[0]):
        if sector_arr[i][0]==ticker_arr[j][0]:
            data={"Ticker":ticker_arr[j][0],"company_name":ticker_arr[j][1],"sector":sector_arr[i][11],"sub sector":sector_arr[i][12],"series":ticker_arr[j][2],"date of listing":ticker_arr[j][3],
            "paid up value":ticker_arr[j][4],"market lot":ticker_arr[j][5],"isin number":ticker_arr[j][6],"face value":ticker_arr[j][7],}
            temp_df=pd.DataFrame(data,index=[k])
            k=k+1
            merge=merge.append(temp_df)
            flag=1
    if flag==0:
        data={"unknown":sector_arr[i][0]}
        temp_df=pd.DataFrame(data,index=[i])
        unknown=unknown.append(temp_df)
```
