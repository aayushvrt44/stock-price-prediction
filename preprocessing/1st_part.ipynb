```python
from telethon.sync import TelegramClient
import datetime as dt
import pandas as pd
import numpy as np
import os
import sys
from asyncio import run
pd.pandas.set_option('display.max_rows',None)
```

```python
df=pd.DataFrame()
df1=pd.DataFrame()
```

```python
api_id="16341762"
api_hash="4981c35b6d5891af091c4d00e97110d2"
```

```python
grps=['t.me/AssetFox']
```

```python
i=0
```

```python
for grp in grps:
    async with TelegramClient("text",api_id,api_hash) as client:
        async for message in client.iter_messages(grp,offset_date=dt.date.today()-dt.timedelta(90), reverse=True):
            #print(message)
            ts = pd.to_datetime(str(message.date)) 
            d = ts.strftime('%Y-%m-%d')
            data={"date":d,"text":message.text}
            temp_df=pd.DataFrame(data,index=[i])
            
            try:
                if(temp_df["text"].dtype=='O' and len(temp_df["text"][i])<40):
                    continue
                else:
                    df=df.append(temp_df)
                    i=i+1
            except:        
                    i=i
```

```python
df = df.dropna(how='any',axis=0)
```
